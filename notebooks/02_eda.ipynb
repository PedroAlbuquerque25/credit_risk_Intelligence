## 1. Libraries & Setup

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv('../data/cs-training-cleaned.csv')

# Preview
df.head()
```

## 2. Target Variable Analysis

```python
df['SeriousDlqin2yrs'].value_counts()
```

```python
# Visualize target variable distribution with percentages
plt.figure(figsize=(8, 5))

ax = sns.countplot(data=df, x='SeriousDlqin2yrs', palette=['#2ecc71', '#e74c3c'])

# Add percentage labels on top of each bar
total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(percentage, (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=12)

plt.title('Default vs Non-Default Clients')
plt.xlabel('0 = Non-Default  |  1 = Default')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## 3. Age Distribution

```python
# Visualize age distribution by target variable
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='SeriousDlqin2yrs', bins=30, kde=True, palette=['#2ecc71', '#e74c3c'])
plt.title('Age Distribution by Default Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Status', labels=['Default', 'Non-Default'])
plt.tight_layout()
plt.show()
```

```python
# Visualize DebtRatio distribution by default status
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='SeriousDlqin2yrs', y='DebtRatio', 
            palette=['#2ecc71', '#e74c3c'])
plt.title('Debt Ratio Distribution by Default Status')
plt.xlabel('0 = Non-Default  |  1 = Default')
plt.ylabel('Debt Ratio')
plt.ylim(0, 2)
plt.tight_layout()
plt.show()
```

```python
# Visualize MonthlyIncome distribution by default status
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='SeriousDlqin2yrs', y='MonthlyIncome', 
            palette=['#2ecc71', '#e74c3c'])
plt.title('Monthly Income Distribution by Default Status')
plt.xlabel('0 = Non-Default  |  1 = Default')
plt.ylabel('Monthly Income')
plt.ylim(0, 10000)
plt.tight_layout()
plt.show()
```

```python
# Visualize NumberOfTime30-59DaysPastDueNotWorse distribution by default status
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='SeriousDlqin2yrs', y='NumberOfTime30-59DaysPastDueNotWorse', 
            palette=['#2ecc71', '#e74c3c'])
plt.title('30-59 Days Past Due Distribution by Default Status')
plt.xlabel('0 = Non-Default  |  1 = Default')
plt.ylabel('Number of Times 30-59 Days Past Due')
plt.ylim(0, 10)
plt.tight_layout()
plt.show()
```

## 5. Correlation Matrix

```python
# Compute and visualize correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', 
            cmap='RdYlGn', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
