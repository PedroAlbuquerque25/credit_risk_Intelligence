# 03 - Feature Engineering
## Credit Risk Intelligence — Canadian Banking Project

## 1. Libraries & Setup

```python
# Import libraries
import pandas as pd
import numpy as np

# Load cleaned dataset
df = pd.read_csv('../data/cs-training-cleaned.csv')
```

## 2. Creating New Features

```python
# Create debt-to-income ratio — how much of income is consumed by debt
df['debt_to_income'] = df['DebtRatio'] * df['MonthlyIncome']

# Create total late payments — sum of all delinquency columns
df['total_late_payments'] = (
    df['NumberOfTime30-59DaysPastDueNotWorse'] +
    df['NumberOfTime60-89DaysPastDueNotWorse'] +
    df['NumberOfTimes90DaysLate']
)

# Create credit utilization flag — high risk if utilization above 75%
df['high_utilization'] = (df['RevolvingUtilizationOfUnsecuredLines'] > 0.75).astype(int)

# Create age groups
df['age_group'] = pd.cut(df['age'], 
                          bins=[20, 30, 40, 50, 60, 110],
                          labels=['20-30', '31-40', '41-50', '51-60', '60+'])
```

```python
# Save the engineered feature set for model training
df.to_csv('../data/cs-training-features.csv', index=False)
```
